```python
import pandas as pd
import pyodbc

def mssql_connection(password: str, username: str, default_db: str = "master"):
                     server = "localhost"
                     
                     cnxn = pyodbc.connect(
                         "DRIVER={ODBC Driver 17 for SQL Server};SERVER="
                         + server
                         + ";DATABASE="
                         + default_db
                         + ";UID="
                         + username
                         + ";PWD="
                         +password
                     )
                     return cnxn
```

```python
conn = mssql_connection(password= "" , username= "" , default_db="summer_2022")
```

```python
import matplotlib.pyplot as plt
plt.close("all")
```

```python
monthly_trips_2020_df = pd.read_sql("SELECT lpep_pickup_datetime, COUNT(*) FROM [2020_taxi_trips] WHERE lpep_pickup_datetime >= '2020-01-01' AND lpep_pickup_datetime <= '2020-12-31' GROUP BY lpep_pickup_datetime ORDER BY lpep_pickup_datetime",conn)
```

```python
monthly_trips_2020_df['lpep_pickup_datetime'] = pd.to_datetime(monthly_trips_2020_df['lpep_pickup_datetime'])
monthly_trips_2020_df = monthly_trips_2020_df.set_index('lpep_pickup_datetime')
```

```python
monthly_trips_2020_df.resample('M').sum()
```

```python
monthly_trips_2020_df.resample('W').sum()
```

```python
import pandas as pd
import pyodbc

def mssql_connection(password: str, username: str, default_db: str = "master"):
                     server = "localhost"
                     
                     cnxn = pyodbc.connect(
                         "DRIVER={ODBC Driver 17 for SQL Server};SERVER="
                         + server
                         + ";DATABASE="
                         + default_db
                         + ";UID="
                         + username
                         + ";PWD="
                         +password
                     )
                     return cnxn
```

```python
conn = mssql_connection(password= "" , username= "" , default_db="summer_2022")
```

```python
weekday_trip_cycle_df = pd.read_sql("SELECT * FROM [2019_taxi_trips] WHERE lpep_pickup_datetime >= '2019-05-13 00:00:00' AND lpep_pickup_datetime <= '2019-05-13 23:59:59' ORDER BY lpep_pickup_datetime",conn)
```

```python
weekday_trip_cycle_df.lpep_pickup_datetime.hist(bins=24, orientation="horizontal");
```

```python
weekend_trip_cycle_df = pd.read_sql("SELECT * FROM [2019_taxi_trips] WHERE lpep_pickup_datetime >= '2019-05-11 00:00:00' AND lpep_pickup_datetime <= '2019-05-11 23:59:59' ORDER BY lpep_pickup_datetime",conn)
```
